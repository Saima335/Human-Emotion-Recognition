```python
# This is an updated version of your code tailored for running on a Kaggle Notebook.
# It uses Kaggle's dataset paths for FER2013 and includes a snippet showing how to
# download the dataset via kagglehub (if needed). 
#
# Note: 
# 1) Kaggle Notebooks often come with popular libraries pre-installed, but if you 
#    need any additional libraries (like kagglehub), you can install them using:
#    !pip install kagglehub
#
# 2) Make sure "msambare/fer2013" is added to your Kaggle Notebook under 
#    "Add data" if you are not using kagglehub. Otherwise, you can comment out
#    the kagglehub part and rely on the Kaggle GUI to link the dataset.


# -----------------------
# 0) (Optional) Download the dataset from Kaggle using kagglehub
# -----------------------
#!pip install kagglehub  # Uncomment if kagglehub is not installed
import kagglehub

# Download latest version of the dataset
path = kagglehub.dataset_download("msambare/fer2013")
print("Path to dataset files:", path)

# -----------------------
# 1) Imports
# -----------------------
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Dense, Dropout, GlobalAveragePooling2D, BatchNormalization
)
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.utils import Sequence

from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import confusion_matrix, classification_report

from tensorflow.keras.layers import Multiply, Conv2D, Activation, Add, Lambda

# -----------------------
# 2) Paths (Kaggle version)
# -----------------------
# After downloading the dataset, Kaggle places it in /kaggle/input/fer2013
train_dir = "/kaggle/input/fer2013/train"
val_dir   = "/kaggle/input/fer2013/test"

batch_size  = 32
num_classes = 7
epochs      = 100  # Adjust as needed

# -----------------------
# 3) Data Generators
#    (Reduced augmentation for speed and stability)
# -----------------------
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25, 
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=[0.9, 1.1],
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    directory=train_dir,
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    directory=val_dir,
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False  # consistent indexing
)

# -----------------------
# 4) Class Weights
# -----------------------
class_indices = train_generator.class_indices
train_classes = train_generator.classes  # numeric labels
classes_unique = np.unique(train_classes)

class_weights_raw = compute_class_weight(
    class_weight='balanced',
    classes=classes_unique,
    y=train_classes
)
class_weights = {i: w for i, w in enumerate(class_weights_raw)}
print("Class Weights:", class_weights)

# -----------------------
# 5) Mixup Sequence
# -----------------------
class MixupSequence(Sequence):
    def __init__(self, base_generator, alpha=0.15):
        self.base_generator = base_generator
        self.alpha = alpha

    def __len__(self):
        # Keras will call this to figure out steps_per_epoch automatically.
        return len(self.base_generator)

    def on_epoch_end(self):
        # Ensure the base generator is reset/shuffled each epoch
        self.base_generator.on_epoch_end()

    def __getitem__(self, index):
        # Acquire one batch from the base_generator
        batch_x, batch_y = self.base_generator[index]

        lam = np.random.beta(self.alpha, self.alpha)
        idx_shuf = np.random.permutation(len(batch_x))

        x_mix = lam * batch_x + (1 - lam) * batch_x[idx_shuf]
        y_mix = lam * batch_y + (1 - lam) * batch_y[idx_shuf]
        return x_mix, y_mix

mixup_seq = MixupSequence(train_generator, alpha=0.15)

# Attention Block
def attention_block(x, filters):
    # Spatial attention: Generate attention map
    attention = Conv2D(filters, (1, 1), activation='sigmoid')(x)  # Generate attention map
    x = Multiply()([x, attention])  # Element-wise multiplication of input and attention map
    return x

# -----------------------
# 6) Define Model
#    (Slightly less dropout)
# -----------------------
def create_efficientnet_model(input_shape=(224, 224, 3), num_classes=7):
    base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=input_shape)

    # Unfreeze the last 40 layers for partial fine-tuning
    for layer in base_model.layers[-40:]:
        layer.trainable = True

    x = base_model.output

    # Attention Block (applied to base model output)
    attention = attention_block(x, filters=x.shape[-1])
    x = GlobalAveragePooling2D()(attention)
    # x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.4)(x)
    x = Dense(256, activation='relu')(x)
    x = Dropout(0.4)(x)
    outputs = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs=base_model.input, outputs=outputs)
    optimizer = Adam(learning_rate=1e-4)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model

model = create_efficientnet_model()
model.summary()

# -----------------------
# 7) Callbacks
# -----------------------
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

model_checkpoint = ModelCheckpoint(
    'best_emotion_model.keras',
    monitor='val_loss',
    save_best_only=True
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=3,
    min_lr=1e-6
)

# -----------------------
# 8) Train
#    We do NOT specify steps_per_epoch or validation_steps,
#    Keras will default to len(mixup_seq) & len(val_generator).
# -----------------------
history = model.fit(
    mixup_seq,
    epochs=epochs,
    validation_data=val_generator,
    class_weight=class_weights,
    callbacks=[early_stop, model_checkpoint, reduce_lr],
    verbose=1
)

# -----------------------
# 9) Evaluate
# -----------------------
val_loss, val_acc = model.evaluate(val_generator, verbose=1)
print(f"\nValidation Accuracy: {val_acc * 100:.2f}%")

# -----------------------
# 10) Confusion Matrix & Classification Report
# -----------------------
val_preds = model.predict(val_generator, verbose=1)
y_true = val_generator.classes
y_pred = np.argmax(val_preds, axis=1)

cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:\n", cm)

class_labels = list(val_generator.class_indices.keys())
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

print("Classification Report:")
print(classification_report(y_true, y_pred, target_names=class_labels))

# -----------------------
# 11) Save Model
# -----------------------
model.save('optimized_emotion_recognition_model.h5')
print("Optimized model saved successfully!")

# -----------------------
# 12) Plot Training Curves
# -----------------------
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

ax[0].plot(history.history['accuracy'], label='Train Accuracy')
ax[0].plot(history.history['val_accuracy'], label='Val Accuracy')
ax[0].set_title('Accuracy')
ax[0].set_xlabel('Epoch')
ax[0].set_ylabel('Accuracy')
ax[0].legend()

ax[1].plot(history.history['loss'], label='Train Loss')
ax[1].plot(history.history['val_loss'], label='Val Loss')
ax[1].set_title('Loss')
ax[1].set_xlabel('Epoch')
ax[1].set_ylabel('Loss')
ax[1].legend()

plt.tight_layout()
plt.show()
```
